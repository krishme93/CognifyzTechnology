### 1.Visualize the locations of restaurants on a map using latitude and longitude information.

```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
df['geometry'] = gpd.GeoSeries([Point(xy) for xy in zip(df['Longitude'], df['Latitude'])])
gdf = gpd.GeoDataFrame(df, geometry='geometry')
gdf.crs = 'EPSG:4326'

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax = world.plot(color='lightgray', edgecolor='black', figsize=(12, 8))
gdf.plot(marker='o', color='red', markersize=5, ax=ax)

ax.set_title("Restaurant Locations Worldwide")
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
plt.show()
```

### 2.Analyze the distribution of restaurants across different cities or countries.

```python
city = df["City"].value_counts().head(20).index.tolist()
count_city = df["City"].value_counts().head(20)
plt.figure(figsize=(15, 10))
sns.countplot(y="City", data=df, order=city, palette='Set1')
plt.title("Distribution of Restaurants by City")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=90)
print(count_city)
plt.show()
```

New Delhi has the highest number of restaurants followed by Gurgaon, Noida and Faridabad in order.

## 3.Determine if there is any correlation between the restaurant's location and its rating.

```python
plt.figure(figsize=(10,6))
correlation = df[["Longitude", "Latitude", "Aggregate rating"]].corr()
sns.heatmap(correlation, linewidth=1, annot=True, cmap="Set1", fmt="2f")
plt.xticks(fontsize=20)
plt.yticks(rotation=90, fontsize=12)
plt.title('Variable Descriptions', fontsize=20)
plt.tight_layout()
plt.show()
```

There is no correlation between Aggregate Rating and Latitude. However, Aggregate Rating and Longitude have a weak negative correlation.
